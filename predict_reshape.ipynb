```python
import pandas as pd
from math import floor,ceil
df = pd.read_csv('predictions.csv')
df = df.loc[:,['ID','item_cnt_month']]
```

```python
df.item_cnt_month = [floor(i) if i-floor(i)<0.4 else i for i in df.item_cnt_month]
df.item_cnt_month = [ceil(i) if ceil(i)-i<0.4 else i for i in df.item_cnt_month]
df.item_cnt_month = [20 if i>20 else i for i in df.item_cnt_month]
df.to_csv('predictions.csv', index=False)
```
